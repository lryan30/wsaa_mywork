**2.** Using the -i option to check the 301 error

-   -i in curl command means include. This tells curl to include the HTTP response headers in the output along with the body of the response. 
-   without the -i flag, curl would only display the content/body of the response not the headers. 
-   By using -i you can see important metadata such as status codes, headers and redirection information. 



HTTP status code 301 means that the requested page has been permanently moved to a new location and the browser/curl is being redirected to a new URL. --> http://www.google.com/.
