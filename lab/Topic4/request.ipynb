```python
import requests
import json

# Define the URL for the API endpoint
url = "http://andrewbeatty1.pythonanywhere.com/books"

def createbook(book):
    # Send a POST request to create the book
    response = requests.post(url, json=book)
    
    # Check if the request was successful
    if response.status_code == 201:  # 201 is the HTTP status for created
        return response.json()  # Return the created book details
    else:
        # Log the error if status code is not 201
        print(f"Error: Received status code {response.status_code}")
        print(f"Response content: {response.text}")
        return None

# Example data for the book we want to create
book_data = {
    "title": "Test Book",
    "author": "Test Author"
}

# Create a new book using the function
created_book = createbook(book_data)

# Print the result of the book creation
if created_book:
    print("Book created successfully:", created_book)
else:
    print("Failed to create book.")
```

```python
# write the update function:

import requests
import json

url = "http://andrewbeatty1.pythonanywhere.com/books"

def updatebook(id, book):
    geturl = url + "/" + str(id)  # Construct the URL for the specific book ID
    response = requests.put(geturl, json=book)  # Send a PUT request to update the book
    return response.json()  # Return the updated book details

```

```python
# write the delete function:
import requests
import json

url = "http://andrewbeatty1.pythonanywhere.com/books"

def deletebook(id):
    geturl = url + "/" + str(id)  # Construct the URL for the specific book ID
    response = requests.delete(geturl)  # Send a DELETE request to delete the book
    return response.status_code  # Return the status code of the response
```
