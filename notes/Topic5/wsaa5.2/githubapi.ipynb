### **How to use authentication to get private information:**



private repository-->aprivateone. -->want to get contents of that repository

-   404-->not found-->not authorised 
-   need to get a key to allow us to do that. 
-   restful API documentation
-   authorisation 
-  get key: 
    -   go to settings
    -   developer settings
    -   personal access token
    -   fine grained token --> can only do limited things
    -   generate new token
    -   if someone has the token they can only access private repo until the date specified. 

Overall, cant access private information from a repository unless you have authorisation from the user and they need to give the key.
