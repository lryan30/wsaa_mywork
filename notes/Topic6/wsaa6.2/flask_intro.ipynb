### **Topic6.2: Flask Introduction**

**Creating your own webservice**

-   will use a package called Flask which will allow you to write the code that will respond to your endpoints for HTTP requests.(postman,curl, python code)
-   Flask lets you write python code that runs on a server and responds to things like visiting a webpage, submitting a form or making an API call.
-   Allows developers to build web applications and APIs quickly and easily. 
-   It provides tools and libraries to handle HTTP requests, URL routing, templates etc.

Flask = App server --> takes the HTTP request from the client (eg: browser or mobile app)and puts together the responses. 
-   can get static files (html,css, javascript, jpeg files etc) and serve it back to the client. like a traditional web server.
-   can also write own code for a web service or app server, like creating a RESTapi, processing from data, link into a database or do some other code linking to a third party(API's, authentication systems)

*Flasks acts as the middle layer between the client and the backend logic wheter its showing a webpage, sending data or running a function*

Principles of all frameworks are the same --> they take in HTTP code requests and give back HTTP responses.
