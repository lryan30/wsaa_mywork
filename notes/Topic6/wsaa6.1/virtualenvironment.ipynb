### **Topic6.1: Virtual Environments:**


*.gitignore*
-   Make a .gitignore file and put the name of the virtual environment directory in it. 
-   This will stop it being uploaded to Github.

*pip-freeze* --> to see all packages on the machine
-  pip freeze > requirements.txt to add all installed packages with versions into a requirements.txt file. 

*pip install -r requirements.txt* --> in the new VM
-   if inside a virtual environment this will capture only packages installed in that environment.
