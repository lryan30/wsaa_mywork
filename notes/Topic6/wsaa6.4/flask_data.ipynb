### **Topic6.4: Flask Data**

How does webserver/flask deal with data
-   how does it receive data that comes up in hTTP request. 
-   how does it give data back to the client/response. 

*From c_Server.py*



*This is a HTTP GET request sent to the flask web server.* 
-   127.0.0.1 is the localhost IP address. it points to own computer. 
-   :5000 means the server is running on port 5000 (flask default)
-   /inquery --> the route/endpoint defined in the flask code
         -      
    -   when someone accesses/inquery, flask looks for the matching function and runs it. 
-   ?name=lou -->query string
    -   comes after the ? in the url
    -   it passes data to the server using key-le value pairs
    -   in this case passing a variable name =lou

1.  User visits /inquery?name=lou
2.  Flask matches the route to this function /inquery
3.  request.args.get('name)--> looks at the query string and pulls out lou
4.  Flask returns the response -->Hello Lou


-   Inbody post requests dont pass data in the url.
-   Instead, they pass data in the body of the request.
-   for form data(login,comments etc), uploading files, sending json(apis)
-   Use postman: 
    -   postman used to test the client side reponse, what the api sends back to the user. 
    -   to see if app is working as expected. 



**Flask returning data:**

-   Rendering Templates vs JSON
    -   Flask can be used to create HTML pages quickly(templates)
    -   not recommended

-   Returning the data(not HTML) and allowing the static web pages use AJAX to get that data
    -   the static webpage (HTML/JS) can then use AJAX(eg with fetch() or jquery) to call the flask API and display data dynamically. 
    -   jsonify() (can return dict object)
    -   return JSON data from flask and let HTML + JS handle how to display it. 

*Note:* Returning a file(img, CSV or PDF etc) is possible with flask using send_file() (outside current scope)



-   Sessions: 
    -   Allow data to be stored between requests (eg: if a user is logged in). 
    -   cookies jwt. 
-   Flashing-->templates, 
-   Logging -->error messages in log file.
