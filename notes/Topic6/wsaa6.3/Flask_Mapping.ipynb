### **Flask Mapping**


Flask mapping refers to the process (mapping) a URL path to a specific function in your code
-   This function is called a view function or a route handler. 
-   This is done using the @app.route() decorator, which tells Flask: 
    -   'When someone visits this URL-->run this function'



HTTP request-->flask will decide if it goes to static page or app-server to give dynamic content back

Flask is event driven. It works by responding to HTTP requests. 
-   Each time you visit a URL (like / or /invalid), 
-   Flask matches the URL to a route (@app/route(....))
-   runs only the function mapped to that route
-   returns the result (eg:a webpage, JSON, redirect etc)
-   then its done --> request/reponse cycle is over. 
-   Redirecting starts a new request, each route runs independently and doesnt contineu on after the redirect. 
-   if want to run code after redirect you cant not in that function but you can write code before the redirect. 
    -   eg: log_user_access()-return redirect(url_for('index'))
    -   pass info using query strings, cookies or session data
    -   chain actions across multiple requests (like login >redirect >profile)

-   can map a particular string, method
-   can also take in the url as a variable in your function
-   if __name__ == '__main__': -- app.run(debug=False)
    -   only runs the flask server if this file is being run directly not if the its being imported into another file. 
-   ```from app import app````
