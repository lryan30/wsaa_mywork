### **Flask Mapping**


Flask mapping refers to the process (mapping) a URL path to a specific function in your code
-   This function is called a view function or a route handler. 
-   This is done using the @app.route() decorator, which tells Flask: 
    -   'When someone visits this URL-->run this function'



HTTP request-->flask will decide if it goes to static page or app-server to give dynamic content back

**Want to create a webservice:**
-   webservice does things depending on what the URL is and can do functionality depending on time or hitting a database
-   gets HTTP requests such as visiting a URL or sending data. 
-   routes the request based on the URL path (eg:login, /weather/today etc)
-   performs actions such as:
    -   accessing or updating a database
    -   running code based on time, user input or other conditions
    -   connecting to external API's
    -   returning dynamic repsonses
    -   Example: you go to /weather/today. the web service checks the current date and weather API. it builds a custom response such as sunny 19C and sends it back. 
-   A webservice reacts to the URL and the type of request, runs code as needed and sends back result.
