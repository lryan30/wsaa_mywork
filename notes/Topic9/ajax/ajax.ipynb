AjAx is a way for your webpage to communicate with the server such as the flask app without refreshing the whole page. 

with Ajax: Javascript can 
-   send a request(get, post, put, delete) to the flask server. 
-   receive the response usually in JSON format.
-   update part of the webpage without a reload. 

*Summary:*

-   can call restful APIs from html/JavaScript pages using AJAX.
