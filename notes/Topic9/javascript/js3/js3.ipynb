DOM:-->document object model. 

-   programming interface HTML and XML documents. 
-   when a webpage loads, the browser creates the DOM->a structured tree of objects representing all the elements like (div, p, button etc) on the page. 
-   DOM lets the programming languages mainly JavaScript access and manipulate the page content, structure and styles dynamically

When you use JavaScript functions like document.getElementById(), document.querySelector(), or change element styles or text — you’re interacting with the DOM.

-   It’s the bridge between the code and what the user sees.



The DOM is a structured representation of the webpage that allows scripts to change whats shown, how it behaves and even the entire structure dynamically.

***How to add elements to the DOM tree:***

The DOM is a programming interface that the browser creates when it loads an HTML page. It turns your HTML into a tree of objects that Javascript can interact with. 

HTML--> DOM Tree --> JavaScript can change it.
