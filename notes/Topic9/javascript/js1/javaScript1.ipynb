Javascript is a programming language that helps make webpages interactive. 
-   it runs behind the scenes to make things like buttons, form submissions or even animations work on websites. 
-   In the example, the function buttonclicked() is written in JavaSCript. 
-   It runs when the button is clicked due to the onclick event.
-   it takes the value from the input box and then updates the <div id ='output'>, 






How do we call the functions when someone clicks on a button on webpage? 

-   onclick attribute of an element. 



<!DOCTYPE html>
<html>
<head>
    <title>Say Hello</title>
</head>
<body>
    <input id="name" type="text" placeholder="Enter your name" />
    <br/>
    <button onclick="ButtonClicked()">Click Me</button>
    <div id="output">Output will go here</div>

    <script>
        function ButtonClicked() {
            var name = document.getElementById("name").value;
            document.getElementById("output").innerText = "Hello," + name + "!";
        }
    </script>
</body>
</html>


-   When the user clicks on the click me button, it runs a function called buttonclicked().
-   output box(div id='output')-->blank area where the message will show after the button is clicked. 
-   script: the function buttonclicked runs when you click the button. 
    -   it gets the name person typed
    -   changes to message to say Hello Name'
