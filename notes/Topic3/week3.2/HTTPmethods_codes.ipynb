**GET:**
-   Purpose: Used to retrieve data from a server (e.g., fetching a webpage or an API endpoint).
-   How it works: Sends data as part of the URL (query parameters).
-   Data visibility: Data is visible in the URL (e.g., ?name=John&age=30).
-   Multiple requests to the same resource should return the same result without causing any side effects.
-   Use: When you want to retrieve information (e.g., searching a website).

**POST:**
-   Purpose: Used to send data to the server to create or update something (e.g., submitting a form or sending data to an API).
-   How it works: Sends data in the body of the request, not in the URL.
-   Data visibility: Data is not visible in the URL (more secure for sensitive data).
-   Repeated requests can create multiple resources (e.g., submitting a form multiple times).
-   Use: When you want to send data to the server, like submitting a login form or uploading a file.


**GET** = Retrieve data

**POST** = Send data (to create or update)
