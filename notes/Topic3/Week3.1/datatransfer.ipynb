### **How to generate a HTTP request?**

-   **Browser:** A browser sends HTTP requests automatically when you navigate to a webpage. (Can view these requests using developer tools under the network tab)

-   **CURL** is a command line tool that allows you to make HTTP requests. For example: you can use the comman ```curl ('url')```

-   **Postman** is a popular GUI-based tool for testing APIs. can create various types of HTTP requests 
(GET,POST etc) and interact with the APIs in a user friendly environment. 

-   **Python**. can use libraries like ```requests``` to generate HTTP requests. For example: ```import requests; response = requests.get(url)```

-   **JavaScript** Uses the ```fetch()``` API or older ```XMLHttpRequest``` to HTTP requests from a web page. ```fetch(url)```

-   **Other languages and programs** Many other programming languages like Java, Ruby, and Go, or tools like HTTPie, can generate HTTP requests using their libraries or tools, offering similar functions to CURL and Postman.

### **CURL**

CURL means client URL. It is a command line tool used to send requests over the internet. It allows you to interact with API's or websites by sending HTTP requests like GET, POST, PUT, DELETE etc. 
-   it is a way to 'talk' to a website or API from the terminal command prompt rather than using a browser or another program. 

```curl -i https://www.atu.ie```

-   give the whole webpage.(v.long)
-   curl-->the command line tool to make HTTP requests
-   -i --> This flag tells curl to include the HTTP response headers in the output (not just the content of the webpage.)
-   HTTP response headers will give you information about the response such as the status code. such as 200 for ok, content type and server details. 
-   HTML content -->the full HTML code of the webpage. 

**POSTMAN**

-   used mostly for API testing and development. 
-   good for debugging
-   easy to use interface for sending HTTP requests (GET, POST, PUT, DELETE etc) to a server and viewing the responses. 
-   particularly useful for developers with RESTful API's 

**How it works:**

Create a Request: You can type in the API URL, choose the request method (e.g., GET, POST), and set any parameters, headers, or body data.

Send the Request: After filling out the necessary details, you hit the "Send" button to make the request to the API.

View Response: Postman will display the response status (e.g., 200 OK), the returned data, and any headers provided by the server
