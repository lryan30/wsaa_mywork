Json is built into Javascript. more compact than xml so more standard for interchanging data across the web. 

similar to dict objects. difference is json/js it must be double quotes, python can have both.

SYNTAX:

-   Name Value Pairs separated by a colon. "name":"Martin"
-   objects identified by curly brackets {}
-   Lists identified by square brckets []
-   All strings (names if space n it) use double quotes not single quotes. 

Example: 



-   dump will dump as a json object
-   dumps as a string
-   load will load in a  json object 
-   loads will load in a string

dumps and loads will parse a string and convert it into a json dict object.
