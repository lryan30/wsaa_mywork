### **Topic 2.1 Representing Data, xml, json, csv.**

When interfacing with APIs with webservers and applications in the cloud,() putting data up and getting data down.)
-   most api's use xml or json. 



**Different ways of storing data:**

-   CSV files: semicolon, comma separated variables. some formats require "", some dont. 
-   SQL database: table, each column has a data type. either an integer or string or image etc
-   XML: human readable text file. it has tags, root tag etc
-   JSON: stores same data but looks different. (similar to printing dict objects-->open/closed squirly brackets {} is an object, open/clsed square brackets are lists [] ). (Name, Value, Pairs)
 

**Attributes to think about different ways of storing data:**

-   Human readable or not
-   How is the data organised (eg:flat-table vs nested) (this matters when you are storing more than one entity type)
-   Ease of modification-updating. sql tables can insert and take tables out etc. 
-   Efficiency of space (especially for transmission on a network). xml takes up more space than json
-   Robustness. 

**XML and JSON:**

-   Human Readable
-   Simple
-   Compact
-   Nested: attributes within attributes
-   Well standardised
