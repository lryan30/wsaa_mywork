db.commit() --> calling db.commit() saves the changes to a database permanently. 
-   if making changes to a database such as insert, update or delete, they are not immediately saved to the actual database file. instead they are staged in a temporary state. calling db.commit() saves the changes permanently. 
-   without db.commit() changes may be lost when connection closed. 
-   if something goes wrong before you commit, you can use db.rollback() to undo the changes.
