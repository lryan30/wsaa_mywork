**DAO: Data Access Object**

design pattern used to separate the data persistance logic(like database queries) from the business logic of an application. 

For example: Instead of writing SQL queries inside your flask routine functions, you write them inside a DAO class. 

What functions would flask call in the DAO?
-   creating another crud API. 


series of python functions you put inside DAO. 
-   flask code will call the functions and will do things to db.
